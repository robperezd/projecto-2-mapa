```python
#import dependencies
import pandas as pd
import numpy as np
import geopandas
import matplotlib.pyplot as plt
```

```python
file_to_load = "Resources/FOMIX_por_Entidad_Federativa_Apoyada_2017_vf.csv"
```

```python
#Read file and store into Pandas data frame
states_data=pd.read_csv(file_to_load)
```

```python
#Print Header
states_data.head()
```

```python
df=pd.DataFrame(states_data)
df
```

```python
df.dtypes
```

```python
Filter_d = df.groupby(["Iniciales","Lat","Lng"])
Filter_DF = (Filter_d["Monto"]).sum().map("$ {:,.0f}".format)
Filter_DF
```

```python
summary_df = pd.DataFrame(Filter_DF).reset_index()
summary_df
```

```python
summary_df.shape
```

```python
summary_df
```

```python
gdf = geopandas.GeoDataFrame(
    summary_df, geometry=geopandas.points_from_xy(summary_df.Lng, summary_df.Lat))
```

```python
gdf
```

```python
gdf.to_file("Resources/gdf.geojson",driver= 'GeoJSON')
```
